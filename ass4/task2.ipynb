Based on the the extended parts list, write a program that creates a dictionary that maps each part number to the quantity and unit of that product.

## Problem Analysis

We have a list with tuples and need to translate it into a dictonary. A dictonary is a { key: value } data type.

The exercise is asking for a dict that uses the part number as key and a tuple of stock number and unit as value.

```python
inventroy = {}

items = [(1, '500-1', 'Hammer', 2, 'Pieces'), (2, '503','Screwdriver', 3, 'Pieces')]

for item in items:
  inventroy[item[2]] = (item[3], item[4])

print(inventroy)



```

Nice we are able to generate the expected result. 

**We solved the exercise.**

*BONUS*: We could also use dict comprehension here.

```python
# Bonus

inventroy2 = {item[2]: (item[3], item[4]) for item in items}
print(inventroy2)
```

**NOTE:** for a dict comprehension follow the following pattern.

variable = { key : value for for_variable in list_input }

Do you see? We need curly braces and a ":" symbol for defining key value pairs.

**Alternative**: We could also try to make the solution more readable.

```python
inventroy3 = { name: (stock, unit) for (_, _, name, stock, unit) in items}

print(inventroy3)
```
