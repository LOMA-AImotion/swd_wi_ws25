## Exercise 1 — List comprehensions II

### (a) Recover 'HELLO' and 'PETER' from mixed letters
The mixed list alternates letters from both words, so we can select by index parity using enumerate().

```python
mixed = ['H','P','E','E','L','T','L','E','O','R']
hello = [ch for i, ch in enumerate(mixed) if i % 2 == 0]
peter = [ch for i, ch in enumerate(mixed) if i % 2 == 1]
hello, peter
```

### (b) Extract product names from parts list
Each tuple is (item, part_no, name, qty, unit). Use list comprehension to get names.

```python
parts = [(1, '500-1', 'Hammer', 2, 'Pieces'), (2, '503', 'Screwdriver', 3, 'Pieces')]
names = [name for (_, _, name, _, _) in parts]
names
```

## Exercise 2 — Dictionaries
Map part number → (quantity, unit) from an extended parts list.

```python
parts = [
    (1, '500-1', 'Hammer', 2, 'Pieces'),
    (2, '503',   'Screwdriver', 3, 'Pieces'),
    (3, '555',   'Nail', 10, 'Pieces'),
]
inventory = {part_no: (qty, unit) for (_, part_no, _name, qty, unit) in parts}
inventory
```

## Exercise 3 — Zip
Storing first and last names in two separate lists is error-prone: indices can get out of sync.
Using tuples (first, last) keeps data together and reduces mismatches. zip() pairs them reliably.

```python
first_names = ['Jane', 'Max', 'Giannis', 'Juan']
last_names  = ['Doe', 'Mustermann', 'Karamitros', 'Perez']
names = list(zip(first_names, last_names))
names
```
