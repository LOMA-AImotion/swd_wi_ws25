## Exercise 1 — List comprehensions II

### (a) Recover 'HELLO' and 'PETER' from mixed letters
The list alternates letters from both words. Even indices belong to 'HELLO', odd to 'PETER'.
Pattern reminder: [expression for (i, ch) in enumerate(mixed) if condition].

```python
mixed = ['H','P','E','E','L','T','L','E','O','R']
# For-loop thinking first
hello_loop = []
peter_loop = []
for i, ch in enumerate(mixed):
    if i % 2 == 0:
        hello_loop.append(ch)
    else:
        peter_loop.append(ch)
hello_loop, peter_loop
```

```python
# List comprehension version
hello = [ch for i, ch in enumerate(mixed) if i % 2 == 0]
peter = [ch for i, ch in enumerate(mixed) if i % 2 == 1]
hello, peter
```

### (b) Extract product names from parts list
Each tuple is (item, part_no, name, qty, unit).
Pattern: [name for (_, _, name, _, _) in parts].

```python
parts = [(1, '500-1', 'Hammer', 2, 'Pieces'), (2, '503', 'Screwdriver', 3, 'Pieces')]
# For-loop first
names_loop = []
for (_item, _no, name, _qty, _unit) in parts:
    names_loop.append(name)
names_loop
```

```python
# List comprehension
names = [name for (_item, _no, name, _qty, _unit) in parts]
names
```

## Exercise 2 — Dictionaries
Map part number → (quantity, unit). Show for-loop then dict comprehension.

```python
parts = [
    (1, '500-1', 'Hammer', 2, 'Pieces'),
    (2, '503',   'Screwdriver', 3, 'Pieces'),
    (3, '555',   'Nail', 10, 'Pieces'),
]
inventory_loop = {}
for (_item, part_no, _name, qty, unit) in parts:
    inventory_loop[part_no] = (qty, unit)
inventory_loop
```

```python
# Dict comprehension
inventory = {part_no: (qty, unit) for (_item, part_no, _name, qty, unit) in parts}
inventory
```

## Exercise 3 — Zip
Storing first and last names separately can get out of sync. Tuples keep related data together.
zip(first_names, last_names) pairs them reliably.

```python
first_names = ['Jane', 'Max', 'Giannis', 'Juan']
last_names  = ['Doe', 'Mustermann', 'Karamitros', 'Perez']
# For-loop first
names_loop = []
for i in range(len(first_names)):
    names_loop.append((first_names[i], last_names[i]))
names_loop
```

```python
# With zip()
names = list(zip(first_names, last_names))
names
```
