# Exercise 3 (Zip, 1 Points)

Imagine that you need to manage the first and last names of people in a program. 

For this purpose you find already existing code, in which all first and last names are stored in separate lists (see example below). 

Briefly explain if this is an optimal solution or if a solution with tuples may be
more appropriate. 

Then use *zip()* to merge the two lists into one list of tuples.


Example: From the lists first names = [’Jane’, ’Max’, ’Giannis’, ’Juan’] and last names
= [’Doe’, ’Mustermann’, ’Karamitros’, ’Perez’] the list names = [(’Jane’, ’Doe’), (’Max’,
’Mustermann’), (’Giannis’, ’Karamitros’), (’Juan’, ’Perez’)] is to be generated.

## Problem analysis

We have two lists (first_names and last_names) and need to combine them to list names containing a tuple of (first_name, last_name).

```python
first_names = ['Jane', 'Max', 'Giannis', 'Juan']
last_names = ['Doe', 'Mustermann', 'Karamitros', 'Perez']
names = zip(first_names, last_names)

print(list(names)) # wrapping it into a list as zip returns an Iterable which will be printed as object
```
