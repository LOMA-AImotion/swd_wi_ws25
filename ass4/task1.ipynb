## a)

Consider the list [’H’, ’P’, ’E’, ’E’, ’L’, ’T’, ’L’, ’E’, ’O’, ’R’]. This list contains the words ’Hello’ and ’Peter’ in mixed form. 

Store the original words in two lists ([’H’, ’E’, ’L’, ’L’, ’O’] and [’P’, ’E’, ’T’, ’E’, ’R’]) using List comprehension.

Tip: Use the enumerate() function.

### Problem analysis

We have a list l that should be splited into two lists l1 and l2. l1 should contain the letters "HELLO" and l2 the letters "PETER".

Is there a pattern that makes it simple to decide if letter n from l should be moved to l1 or l2? And how could enumerate() help us here?


```python
l = ['H', 'P', 'E', 'E', 'L', 'T', 'L', 'E', 'O', 'R']

for index, entry in enumerate(l):
    print(f"index: {index} : letter: {entry}")
```

We can clearly see that we could use the index (odd or even) to decide if something should be placed in l1 or l2.

```python
l1 = []
l2 = []

for index, entry in enumerate(l):
    if index % 2 == 0:
        l1.append(entry)
    else:
        l2.append(entry)

print(f"l1: {l1} : l2: {l2}")
```

We have a working solution but we are not done yet. The exercise clearly askes for a **list comprehension**. But with the simple for loop in place the translation should be simple for us.

```python
l1 = [entry for index, entry in enumerate(l) if index % 2 == 0]
l2 = [entry for index, entry in enumerate(l) if index % 2 != 0]

print(f"l1: {l1} : l2: {l2}")
```

Imagine a parts list that lists the following information for each product: Item, part number,name, quantity and unit. 

Extract the names of all products from a list of tuples with all the information listed. Use List comprehension to do this.

Example: From the list parts = [(1, ’500-1’, ’Hammer’, 2, ’Pieces’), (2, ’503’,’Screwdriver’, 3, ’Pieces’)] the list [’Hammer’, ’Screwdriver’] shall be extracted.

## Problem analysis

We have a item list ( lets call it items) that contains tuples that represent parts providing information about the index, partnumber, name, stock number and unit.

**We already learned that the index of a tuple can become handy in such situations.**


```python
items = [(1, '500-1', 'Hammer', 2, 'Pieces'), (2, '503','Screwdriver', 3, 'Pieces')]

for item in items:
    print(f"Name: {item[2]}, full info: {item}")
```

Ok halfway trough, now that we understand the data structure we need to translate this into a **list comprehension** to fullfill the exercise.

```python
names = [item[2] for item in items]
print(names)
```
