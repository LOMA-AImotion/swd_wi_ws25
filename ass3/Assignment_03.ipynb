## Exercise 1 — While-loops (min, max, mean)
Enter integers until 'stop', then compute statistics using a for-loop.

```python
data = []
while True:
    raw = input("Enter an integer or 'stop': ").strip()
    if raw.lower() == 'stop':
        break
    data.append(int(raw))

# Compute min, max, mean using a for-loop
mn = data[0]
mx = data[0]
total = 0
count = 0
for x in data:
    if x < mn: mn = x
    if x > mx: mx = x
    total += x
    count += 1
mean = total / count
print('min:', mn, 'max:', mx, 'mean:', mean)
```

## Exercise 2 — Loop conversions
We show both sides to understand equivalence.

### (a) List comprehension → for-loop
Original list comprehension collects last names.

```python
customers = [
    ['Max', 'Mustermann', '01.01.83'],
    ['Martina', 'Musterfrau', '02.02.84'],
    ['Gabi', 'Meier', '03.03.85'],
]
# List comprehension
lc_last_names = [person[1] for person in customers]

# For-loop equivalent
fl_last_names = []
for person in customers:
    fl_last_names.append(person[1])
lc_last_names, fl_last_names
```

### (b) for-loop → while-loop
Square a list of numbers using while.

```python
numbers = [1, 4, 2, 8, 5]
squared_numbers = []
i = 0
while i < len(numbers):
    squared_numbers.append(numbers[i] * numbers[i])
    i += 1
squared_numbers
```

### (c) while-loop → for-loop
We can write the guessing interaction with a for-loop by iterating over inputs until we match the secret.

```python
import random
secret_number = random.randint(1, 100)
for guess in iter(lambda: int(input('Guess a number between 1 and 100: ')), secret_number):
    if guess > secret_number:
        print('Too high! Guess again.')
    elif guess < secret_number:
        print('Too low! Guess again.')
print('You guessed it! The number was', secret_number)
```

## Exercise 3 — Word count
We first do it with a for-loop and list.count as hinted; then show a shorter variant.

```python
s = 'I study at THI and I have fun'
words = s.split()
unique_words = set(words)
counts = {}
for w in unique_words:
    counts[w] = words.count(w)
counts
```

```python
# Optional shorter (dict comprehension):
{w: words.count(w) for w in unique_words}
```
