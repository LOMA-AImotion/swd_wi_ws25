## Exercise 1 — While-loops (min, max, mean)
Enter integers until 'stop', then compute statistics using a for-loop.

```python
def stats_min_max_mean(nums: list[int]) -> dict:
    # Assume nums is non-empty
    mn = nums[0]
    mx = nums[0]
    total = 0
    count = 0
    for x in nums:
        if x < mn: mn = x
        if x > mx: mx = x
        total += x
        count += 1
    mean = total / count
    return {'min': mn, 'max': mx, 'mean': mean}

# Demo with example sequence: 3, 7, 2
stats_min_max_mean([3, 7, 2])
```

## Exercise 2 — Loop conversions

### (a) List comprehension → for-loop
Original list comprehension collects last names.

```python
customers = [
    ['Max', 'Mustermann', '01.01.83'],
    ['Martina', 'Musterfrau', '02.02.84'],
    ['Gabi', 'Meier', '03.03.85'],
]
# List comprehension
lc_last_names = [person[1] for person in customers]

# For-loop equivalent
fl_last_names = []
for person in customers:
    fl_last_names.append(person[1])
lc_last_names, fl_last_names
```

### (b) for-loop → while-loop
Square a list of numbers using while.

```python
numbers = [1, 4, 2, 8, 5]
squared_numbers = []
i = 0
while i < len(numbers):
    squared_numbers.append(numbers[i] * numbers[i])
    i += 1
squared_numbers
```

### (c) while-loop → for-loop
Use iter(callable, sentinel) to stop when the guessed number equals the secret. This avoids an unbounded for-range.

```python
import random
secret_number = random.randint(1, 100)
# The loop runs until input equals secret_number; then the for-loop finishes and we print the success message.
for guess in iter(lambda: int(input('Guess a number between 1 and 100: ')), secret_number):
    if guess > secret_number:
        print('Too high! Guess again.')
    elif guess < secret_number:
        print('Too low! Guess again.')
print('You guessed it! The number was', secret_number)
```

## Exercise 3 — Word count
Count each word's frequency using list.count (as hinted).

```python
s = 'I study at THI and I have fun'
words = s.split()
unique = set(words)
counts = {w: words.count(w) for w in unique}
counts
```
