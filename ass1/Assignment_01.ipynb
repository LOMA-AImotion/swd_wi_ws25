## Exercise 1 — Geometric formulas
We compute the area of a rectangle and a simple price breakdown with tax.

### How to think
- Identify the given values (length, width, price per m², tax).
- Compute area = length × width.
- Net price = area × price per m²; Tax = net × 0.035; Gross = net + tax.
Start with fixed numbers, then make it interactive.

```python
# (a) Naive calculation: area of 4m x 3m
area = 4 * 3
area
```

```python
# (b) Using variables with meaningful names
length_m = 4
width_m = 3
area_sqm = length_m * width_m
area_sqm
```

```python
# (c) Price calculation with 3.5% tax — step by step (no custom functions)
price_per_sqm = 50.0
tax_rate = 0.035
net = area_sqm * price_per_sqm
tax = net * tax_rate
gross = net + tax
net, tax, gross
```

```python
# Optional: make it interactive using input()
length_m = float(input('Enter length (m): ').replace(',', '.'))
width_m = float(input('Enter width (m): ').replace(',', '.'))
price_per_sqm = float(input('Enter price per m² (€): ').replace(',', '.'))
area_sqm = length_m * width_m
net = area_sqm * price_per_sqm
tax = net * 0.035
gross = net + tax
print('Area (m²):', area_sqm)
print('Net (EUR):', net)
print('Tax 3.5% (EUR):', tax)
print('Gross (EUR):', gross)
```

## Exercise 2 — Dog years
Rules: 0→0, 1→14, 2→22, each additional dog year adds +5 human years.
We first write it with simple if/elif/else and input().

```python
dog_years = int(input('Dog age (years): '))
if dog_years == 0:
    human_years = 0
elif dog_years == 1:
    human_years = 14
elif dog_years == 2:
    human_years = 22
else:
    human_years = 22 + 5 * (dog_years - 2)
human_years
```

You can also inspect several values quickly using a for-loop (no functions yet):

```python
table = []
for y in range(0, 9):
    if y == 0:
        h = 0
    elif y == 1:
        h = 14
    elif y == 2:
        h = 22
    else:
        h = 22 + 5 * (y - 2)
    table.append((y, h))
table
```

## Exercise 3 — Leap years
Rule: leap if divisible by 4, except if divisible by 100 (not leap), unless divisible by 400 (leap).

```python
year = int(input('Enter a year: '))
if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(year, 'is a leap year')
else:
    print(year, 'is not a leap year')
```
