## Exercise 1 — Triangles
Print a triangle with a chosen character and height.

### How to think
- Draw it on paper: lines grow from 1 to n, then shrink from n-1 back to 1.
- Translate that into two for-loops.

```python
symbol = input('Symbol: ') or '#'
n = int(input('Height: '))
# Upward part
for i in range(1, n + 1):
    print(symbol * i)
# Downward part
for i in range(n - 1, 0, -1):
    print(symbol * i)
```

## Exercise 2 — List operations
Ask for k words and compute the total number of characters using a for-loop.

```python
k = int(input('How many words? '))
words = []
for i in range(k):
    w = input(f'Word {i+1}: ')
    words.append(w)
total = 0
for w in words:
    total += len(w)
print('Total length:', total)
```

Optional (shorter): Python has built-ins like sum() and len().
We can write the same as: sum(len(w) for w in words)

```python
sum(len(w) for w in words)
```

## Exercise 3 — List comprehensions (only list comprehension)
We first show the for-loop version (thinking process), then the equivalent list comprehension.

General pattern: [expression for item in iterable if condition] (the if-part is optional).

### (a) Repeated character list

```python
# For-loop first
char = '#'
n = 3
repeats_loop = []
for i in range(1, n + 1):
    repeats_loop.append(char * i)
repeats_loop
```

```python
# List comprehension
repeats_comp = [char * i for i in range(1, n + 1)]
repeats_comp
```

### (b) Multiply number by factors

```python
number = 3
factors = [1, 2, 3]
products_loop = []
for f in factors:
    products_loop.append(number * f)
products_loop
```

```python
products_comp = [number * f for f in factors]
products_comp
```

### (c) Split into even/odd

```python
data = [1, 2, 3]
evens_loop = []
odds_loop = []
for x in data:
    if x % 2 == 0:
        evens_loop.append(x)
    else:
        odds_loop.append(x)
evens_loop, odds_loop
```

```python
evens_comp = [x for x in data if x % 2 == 0]
odds_comp  = [x for x in data if x % 2 != 0]
evens_comp, odds_comp
```
