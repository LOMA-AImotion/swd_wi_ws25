## Exercise 1 — Triangles
Print a triangle with a chosen character and height.

```python
def print_triangle(symbol: str, n: int) -> None:
    # Upward part
    for i in range(1, n + 1):
        print(symbol * i)
    # Downward part
    for i in range(n - 1, 0, -1):
        print(symbol * i)

# Example
print_triangle('#', 3)
```

## Exercise 2 — List operations
Ask for k words and compute the total number of characters using a for-loop.

```python
def total_length(words: list[str]) -> int:
    total = 0
    for w in words:
        total += len(w)
    return total

# Example
words = ['THI', 'EGM']
total_length(words)
```

## Exercise 3 — List comprehensions (only list comprehension)
(a) Repeated character list, (b) multiply number by factors, (c) split into even/odd.

```python
# (a)
char = '#'; n = 3
repeats = [char * i for i in range(1, n + 1)]
repeats
```

```python
# (b)
number = 3
factors = [1, 2, 3]
products = [number * f for f in factors]
products
```

```python
# (c)
data = [1, 2, 3]
evens = [x for x in data if x % 2 == 0]
odds  = [x for x in data if x % 2 != 0]
evens, odds
```
