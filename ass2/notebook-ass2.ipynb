# Exercise 1 (Triangles, 1 Points)

Write a program that prints a triangle on the command line from a character of your choice with
a height entered by the user.

Example: For user input 3, the program should produce an output similar to this:
```bash 
#
##
###
##
#
```

## Analyzing the problem

-> The shape build by x rows, containing y repetitions of "#"

1. We need to find x (number of rows)
2. We need to find to now y ( number of repetitions of '#')
    2.1 We now the at first we increase the number of '#'
    2.2 The row in the middle contains the most '#' which seems to be n ( the input)
    2.3 Then the repetitions get less until we reach a single '#'

Viewing it mathmatically:

For n = 3 we can observe that the shape will have 5 rows.

```bash 
1: #
2: ##
3: ###
4: ##
5: #
```
For n = 5 we can observe that the shape will have 5 rows.

```bash
1: #
2: ##
3: ###
4: ####
5: #####
6: ####
7: ###
8: ##
9: #
```

Can we see a pattern there? Some kind of formular?

Yes it seems like the number of rows (x) could be calculated by x = n * 2 - 1 lets check this

`n=3; x = 3 * 2 - 1 result: x = 5 `

`n=5; x = 5 * 2 - 1 result: x = 9`

Looks solid! **Lets code this!**


````python
n = 3
number_of_rows = n * 2 - 1

print(f"Number of rows required for n={n}: rows={number_of_rows}")
````

Nice, the calculation is solid, now lets place the rows.

````python
# First we need a range we could write something like:
rows = range(number_of_rows)
print(rows)

# and loop over it like:
for index in rows:
    print(f"Row: {index}")
````

Technically that is all we need, by this we get 4 rows but we are starting at 0 as we learned this is convention in computer science.
This can be confusing when talking about rows, we could correct this by doing this: 

````python
for index in rows:
    print(f"Row: {index + 1}")
````

But correcting the index manually can be complicated and hard to track more simple is it to use the range(a, b) function insted of the range(b) function. 
```
range(b): creates a range starting from 0 to b
range(a, b): crates a range starting from a to b
```

````python
rows = range(number_of_rows)
print(rows)
rows = range(1, number_of_rows + 1) # Think about why we need the +1 here!
print(rows) 

for index in rows:
    print(f"Row: {index}")
````

Did you spot the difference? Now things get very simple for the first half: 


````python
for index in rows:
    print(f"row-{index}: {index * '#'}")
````

So row 1 to 3 seems good, but as we can se after row 4 we need to reduce the number of '#' symbols with each iterations.

But we can solve this with math!


````python
for index in rows:
    if index <= n:
        print(index * '#')
    else:
        print("#" * (n - (index % n)))
````

````python
symbol = "#"

n = int(input("Size of triangle? "))

for index in range(n * 2 - 1):
    if index <= n - 1:
        print(symbol * (index + 1))
    else:
        print(symbol * (n - (index % n + 1)))
````
